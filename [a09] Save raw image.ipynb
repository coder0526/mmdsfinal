```python
import ACMdev, csv
root = ACMdev.TsvData['root']
```

```python
session = ACMdev.getSession()
with open(ACMdev.TsvData['trnImage'],'rb') as tsvin:
    tsvin = csv.reader(tsvin, delimiter='\t')
    p = ACMdev.ProgressBar(1)
    j = 0
    for row in tsvin:
        saveImage(j, row[1])
        data = ACMdev.TrnImage( j, row[0])
        session.add(data)
        if j % 50 == 0: session.commit()
        if j % 5000 == 0: p.showStep(j)
        j= j + 1

    session.commit()
session.close()
print "Done!"
```

```python
session = ACMdev.getSession()
with open(ACMdev.TsvData['trnLog'],'rb') as tsvin:
    tsvin = csv.reader(tsvin, delimiter='\t')
    p = ACMdev.ProgressBar(1)
    j = 0
    for row in tsvin:
        data = ACMdev.TrnLog(row[0], row[1], int(row[2]))
        session.add(data)
        if j % 50 == 0: session.commit()
        if j % 5000 == 0: p.showStep(j)
        j= j + 1

    session.commit()
session.close()
print "Done!"
```

```python
def saveImage(id, data):
    f = open(root + str(id) + ".png", "w")
    f.write(data.decode('base64'))
    f.close()
    
```
